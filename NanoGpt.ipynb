```python
with open('Shakespeare.txt','r',encoding='utf-8') as f:
  text=f.read()
  print(text[:1000])
```

```python
sampleset_chars=sorted(list(set(text)))
print(sampleset_chars,len(sampleset_chars))
vocab_size=len(sampleset_chars)
```

```python
import torch
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
#My code - Char to Num
def cton(ch):
    print(ch)
    for t in list(enumerate(sampleset_chars)):
      if(ch==t[1]):
        return t[0]

char_to_num= lambda s:list([cton(ch) for ch in s])
char_to_num(text[:1000])

itos = { i:ch for i,ch in enumerate(sampleset_chars) }
decode = lambda l: ''.join([itos[i] for i in l])

# for sample_char in sample_chars
# char_to_num(text[:1000])
```

```python
import torch
data=torch.tensor(char_to_num(text),dtype=torch.long)
data[:100]
```

```python
n = int(0.9*len(data))
train_data = data[:n]
val_data = data[n:]
```

```python
block_size=8

x_train= train_data[:block_size]
y_train=train_data[1:block_size+1]
for i in range(block_size):
  context= x_train[:i+1]
  target=y_train[i]
  print(f"when input is {context} the target: {target}")
```

```python
torch.manual_seed(1337)
batch_size = 4 # how many independent sequences will we process in parallel?
block_size = 8 # what is the maximum context length for predictions?

def get_batch(split):
    # generate a small batch of data of inputs x and targets y
    data = train_data if split == 'train' else val_data
    ix = torch.randint(len(data) - block_size, (batch_size,))
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i+1:i+block_size+1] for i in ix])
    x,y=x.to(device),y.to(device)
    return x, y

xb, yb = get_batch('train')
print('inputs:')
print(xb.shape)
print(xb)
print('targets:')
print(yb.shape)
print(yb)

print('----')

for b in range(batch_size): # batch dimension
    for t in range(block_size): # time dimension
        context = xb[b, :t+1]
        target = yb[b,t]
        # print(f"when input is {context.tolist()} the target: {target}")
```

```python
import torch
import torch.nn as nn
from torch.nn import functional as F

n_embd=32
class BigramLanguageModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.token_embedded_table= nn.Embedding(vocab_size,n_embd)
    self.position_embedded_table=nn.Embedding(block_size,n_embd)
    self.lm_head=nn.Linear(n_embd,vocab_size)

  def forward(self,xb,yb=None):
    B,T=xb.shape
    tok_emd=self.token_embedded_table(xb)
    # Only consider the last block_size tokens for position embeddings
    pos_emd=self.position_embedded_table(torch.arange(T,device=device))
    # print(pos_emd.shape)
    x=tok_emd+pos_emd
    logits=self.lm_head(x)
    if yb == None:
      loss = None
    else:
      B,T,C=(logits.shape)
      logits=logits.view(B*T,C)
      yb=yb.view(B*T)

      loss=F.cross_entropy(logits,yb)
    return logits,loss



  def generate(self,xb,pred_n_tokens):
    # Add device argument to torch.zeros
    xb = xb.to(device)
    for _ in range(pred_n_tokens):
      # Crop xb to the last block_size tokens if it's longer
      xb_cropped = xb[:, -block_size:] if xb.shape[1] > block_size else xb
      logits,loss=self(xb_cropped)
      next_token=logits[:,-1,:]
      next_token = next_token - next_token.max(dim=-1, keepdim=True).values
      probs = F.softmax(next_token/1.0, dim=-1) # (B, C)
      probs = torch.nan_to_num(probs, nan=0.0)
      idx_next = torch.multinomial(probs, num_samples=1) # (B, 1)
      xb = torch.cat((xb, idx_next), dim=1) # (B, T+1)
    return xb

model=BigramLanguageModel().to(device)
pred,loss=model(xb,yb)
print(pred.shape)
print(loss)

# print(model.generate(torch.zeros((1,1),dtype=torch.long),pred_n_tokens=100).tolist())
```
